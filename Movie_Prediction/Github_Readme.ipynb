# üé¨ IMDB Movie Rating Predictor (Metadata-Based Regression)

This project focuses on predicting a movie's **aggregate IMDB Rating** by analyzing its intrinsic features such as box office revenue, critic scores, runtime, and genre. While the initial goal was user-based collaborative filtering, the provided dataset necessitated a shift to a robust regression model to determine the core factors influencing a film's public success.

| Metric | Value | Interpretation |
| :--- | :--- | :--- |
| **Model** | Random Forest Regressor | Used for non-linear prediction and feature analysis. |
| **RMSE** (Root Mean Squared Error) | **0.4312** | The average error in prediction is only about 0.43 points on the 10-point rating scale. |
| **R-squared ($\mathbf{R}^2$)** | **0.7410** | Approximately 74.1% of the variance in IMDB Ratings is explained by the model's features. |

## üåü Key Insights (What Drives a Movie's Rating?)

The Feature Importance analysis revealed that the following features are most critical in predicting a movie's final rating:

1.  **Votes**: The sheer volume of public participation (number of votes) is the single most important factor, likely signifying sustained public interest and a stabilized score.
2.  **Metascore**: The critical consensus (Metascore) is highly correlated with the public average, making it the second most predictive feature.
3.  **Revenue (Millions)**: Financial success is a significant, but secondary, indicator of quality.

## üõ†Ô∏è Methodology

1.  **Data Preprocessing**: Handled missing values in the `Revenue (Millions)` and `Metascore` columns via list-wise deletion.
2.  **Feature Engineering**: The multi-valued `Genre` column was transformed using **One-Hot Encoding** for the top 10 most frequent genres.
3.  **Model Selection**: A **Random Forest Regressor** was trained to capture complex, non-linear relationships between the movie metadata and the target `Rating`.
4.  **Evaluation**: The model was validated on a 20% test set using RMSE and $\text{R}^2$ metrics.
